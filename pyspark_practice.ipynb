```python
# create a spark session to use Spark

from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Practice").getOrCreate()
```

```python
# read csv, with header on and inferSchema on
# to get correct types
df_pyspark = spark.read.csv('test2.csv', header=True, inferSchema=True)
```

```python
# shows dataframe
df_pyspark.show()
```

```python
# removes column
df_pyspark.drop('Name').show()
```

```python
df_pyspark.show()
```

```python
# remove all rows with null values in any column
df_pyspark.na.drop().show()
# df_pyspark.na.drop(how="any").show()
```

```python
# remove all rows with all columns are null values
df_pyspark.na.drop(how="all").show()
```

```python
df_pyspark.na.drop(how="any", thresh=2).show()
# drop rows with less than 2 non-null value
```

```python
# remove null values for a specific column (here Age)

df_pyspark.na.drop(how="any", subset=["Age"]).show()
```

Part III

```python
df_pyspark = spark.read.csv('test2.csv', header=True)
```

```python
df_pyspark.show()
```

```python
# replace na value with new label
df_pyspark.na.fill('Missing Values').show()
```

```python
# only replace na value for specific column(s)
df_pyspark.na.fill('Missing Value', ['Experience', 'age']).show()
```

```python
df_pyspark.show()
```

```python
# replace null value with column average
# ex. experience

# using imputer function, which automatically calculates column
# average and replaces null values
from pyspark.ml.feature import Imputer

imputer = Imputer(
    inputCols = ['Age', 'Experience', 'Salary'],
    outputCols = ["{}_imputed".format(c) for c in ['Age', 'Experience', 'Salary']]
).setStrategy('mean')
```

```python
df_pyspark = spark.read.csv("test2.csv", header=True, inferSchema=True)
df_pyspark.show()
```

```python
imputer.fit(df_pyspark).transform(df_pyspark).show()
```

### Part IV

```python
# setup and create SparkSession
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Practice").getOrCreate()
```

```python
df_pyspark = spark.read.csv('test3.csv', header=True, inferSchema=True)
df_pyspark.show()
```

```python
df_pyspark.printSchema()
```

```python
# filter: salary of people <= 20000
# filter function uses string
# or specify using DF and column
df_pyspark.filter("Salary<=20000").select(['Name', 'Age'])

df_pyspark.filter((df_pyspark['Salary'] <= 20000) | (df_pyspark['Salary'] >= 15000)).show()
```

```python
# use not (~) to negate filter (> 20000)

df_pyspark.filter(~(df_pyspark['Salary'] <= 20000)).show()
```

### Part V

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Agg').getOrCreate()
```

```python
spark
```

```python
df_pyspark = spark.read.csv("test4.csv", header=True, inferSchema=True)
```

```python
df_pyspark.show()
```

```python
df_pyspark.printSchema()
```

```python
# group by function
# group by + aggregate works together (here .sum)
df_pyspark.groupBy('Name').sum('Salary').show()
```

```python
# find which department gives max salary
df_pyspark.groupBy('Departments').sum('Salary').show()
```

```python
# department average salaries
df_pyspark.groupBy('Departments').mean('Salary').show()
```

```python
# get counts of each department
df_pyspark.groupBy('Departments').count().show()
```

```python
# aggregate for a single column
df_pyspark.agg({'Salary': 'sum'}).show()
```

```python
# who is getting max salary
df_pyspark.groupBy('Name').max('Salary').show()
```
